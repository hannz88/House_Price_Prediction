```python
# encode the string into int
# DecisionTree does not take "string"
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
d = defaultdict(LabelEncoder)
train_features_encoded = train_features.apply(lambda x: d[x.name].fit_transform(x))
# this is one way to do it but it will transform all columns
```

```python
# another to transform them is by columns
from sklearn.preprocessing import LabelEncoder
labelencode = LabelEncoder()
train_features["Neighbour_encoded"] = labelencode.fit_transform(train_features["Neighborhood"])
train_features["condition1_encoded"] = labelencode.fit_transform(train_features["Condition1"])
train_features
```

```python
# remove the original columns
train_features = train_features.drop(["Neighborhood", "Condition1"], axis=1)
train_features.head()
```

```python
# split the train data into validation and actual training data
from sklearn.model_selection import train_test_split
subtrain_features, val_features, subtrain_price, val_price = train_test_split(train_features, train_price, random_state=1)
```

# Decision Tree Regressor

```python
# use DecisionTree on train data
from sklearn.tree import DecisionTreeRegressor
house_price_model = DecisionTreeRegressor(random_state = 1)  # build a model
```

```python
house_price_model.fit(subtrain_features, subtrain_price)  # fit the model
```

```python
# predict and calculate mean absolute error
predicted_price = house_price_model.predict(val_features)
# mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(predicted_price, val_price)
mae
```

```python
# optimise max_leaf_nodes and reduce the mae
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# find the optimal max_leaf_nodes
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, subtrain_features, val_features, subtrain_price, val_price)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
# 50 is the most optimal max_leaf_nodes
```

# Random Tree Regressor

```python
# set up the model
from sklearn.ensemble import RandomForestRegressor
rt_model = RandomForestRegressor(random_state=1)  # set up the model
rt_model.fit(subtrain_features,subtrain_price)  # train the model
rt_predicted_price = rt_model.predict(val_features)  # predict the validation price
rt_mae = mean_absolute_error(rt_predicted_price, val_price)  # compute the mae
rt_mae
```

```python
# optimise max_leaf_nodes for random forest
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# find the optimal max_leaf_nodes
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, subtrain_features, val_features, subtrain_price, val_price)
    print("Max leaf nodes in Random Forest: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
# 50 is the most optimal max_leaf_nodes
```

Random Tree Regressor is the better model here and the most optimal number of max_leaf_nodes for both are 50. RandomTreeRegressor is chosen to predict the price of the test data.
